```python
import pandas as pd
import numpy as np 
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from preprocess import preprocess_text 
df=pd.read_csv("../../data/final_df/quali_final_df2.csv")
nlp = spacy.load('en_core_web_lg')
```

```python
#Parse the text column using spacy
df['text_original'] = df['text']
df['text_parsed'] = df['text'].apply(nlp)
df['text_parsed'] = df['text_parsed'].apply(lambda doc: ' '.join([token.text for token in doc]))#unlist/unspacy

columns_to_preprocess = ['text', 'lag_text','lag_text2', 'lag_text3']

for column in columns_to_preprocess:
    df[column] = df[column].apply(lambda x:preprocess_text(x, nlp))
```

```python
X = df[["text", 'previous_column_number', 'str_len', "lag_text", 'lag_text2', 'lag_text3',  'entity_encode']]
```

```python
df.dtypes
# Assuming df is your second dataframe
df['dsm'] = df['dsm'].astype('object')
df['entity_encode'] = df['entity_encode'].astype('object')
df.dtypes
```

```python
from joblib import load
ps_preprocessor2 = load('preprocessors/quali_processor.joblib')

matrix_x = ps_preprocessor2.transform(X)
```

```python
qua_model = load('../../joblib/qualifications_rfm.joblib')
predictions = qua_model.predict(matrix_x)
df['predictions_model1'] = predictions
```

```python
# Get the probabilities of the positive class
# probabilities = ps_model.predict_proba(matrix_x)[:, 1]
probabilities = qua_model.predict_proba(matrix_x)
for i, class_name in enumerate(qua_model.classes_):
    df[f'model1_probability_class_{class_name}'] = probabilities[:, i]
```

```python
df.to_csv('../../data/final_df/quali_classified.csv', index=False)
```
